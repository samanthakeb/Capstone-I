```python
import pandas as pd

life_expectancy = pd.read_csv('C:\\Users\\saman\\OneDrive\\Documents\\Coding Temple Work\\lifeexpectancy.csv')

print(life_expectancy)
```

```python
selected_countries = ["Albania", "Algeria", "Argentina", "Australia", "Austria", "Bahamas", "Belgium", "Bhutan", "Botswana", "Brazil", "Bulgaria", "Burkina Faso", "Canada", "Chile", "China", "Colombia", "Costa Rica", "Croatia", "Cuba", "Czech Republic", "Denmark", "Egypt", "Finland", "France", "Georgia", "Germany", "Ghana", "Greece", "Hong Kong", "Iceland", "India", "Indonesia", "Ireland", "Israel", "Italy", "Japan", "Kuwait", "Liechtenstein", "Luxembourg", "Macau", "Malaysia", "Maldives", "Mauritius", "Mexico", "Monaco", "Morocco", "Netherlands", "New Zealand", "North Korea", "Norway", "Pakistan", "Peru", "Phillipines", "Portugal", "Romania", "Russia", "Rwanda", "Serbia", "Seychelles", "Singapore", "South Africa", "South Korea", "Spain", "Sri Lanka", "Suriname", "Sweden", "Switzerland", "Taiwan", "Thailand", "Trinidad and Tobago", "Tunisia", "Turkey", "United Kingdom"]

universal_healthcare_data = life_expectancy[life_expectancy['Country'].isin(selected_countries)]

print(universal_healthcare_data)
```

```python
life_expectancy['Universal.Healthcare'] = life_expectancy['Country'].isin(universal_healthcare_data['Country'])

print(life_expectancy)
```

```python
import matplotlib.pyplot as plt

females_with_uh = life_expectancy[(life_expectancy['Year'] >= 2000) & 
                                   (life_expectancy['Year'] <= 2020) & 
                                   (life_expectancy['Universal.Healthcare'] == True)]

female_average_uh = females_with_uh.groupby('Year')['Female Life Expectancy'].mean().reset_index()

print(female_average_uh)

plt.plot(female_average_uh['Year'], female_average_uh['Female Life Expectancy'])
plt.title("Average Female Life Expectancy in Countries with Universal Healthcare 2000-2020")
plt.xlabel("Year")
plt.ylabel("Average Female Life Expectancy")
plt.show()
```

```python
females_without_uh = life_expectancy[(life_expectancy['Year'] >= 2000) & 
                                      (life_expectancy['Year'] <= 2020) & 
                                      (life_expectancy['Universal.Healthcare'] == False)]


female_average = females_without_uh.groupby('Year')['Female Life Expectancy'].mean().reset_index()

print(female_average)

plt.plot(female_average['Year'], female_average['Female Life Expectancy'])
plt.title("Average Females Life Expectancy in Countries without Universal Healthcare 2000-2020")
plt.xlabel("Year")
plt.ylabel("Average Female Life Expectancy")
plt.show()
```

```python
males_with_UH = life_expectancy[(life_expectancy['Year'] >= 2000) & 
                                (life_expectancy['Year'] <= 2020) & 
                                (life_expectancy['Universal.Healthcare'] == True)]

male_average_with_UH = males_with_UH.groupby('Year')['Male Life Expectancy'].mean().reset_index()

print(male_average_with_UH)

plt.plot(male_average_with_UH['Year'], male_average_with_UH['Male Life Expectancy'])
plt.title("Average Male Life Expectancy in Countries with Universal Healthcare (2000-2020)")
plt.xlabel("Year")
plt.ylabel("Average Male Life Expectancy")
plt.show()
```

```python
males_no_UH = life_expectancy[(life_expectancy['Year'] >= 2000) & 
                               (life_expectancy['Year'] <= 2020) & 
                               (life_expectancy['Universal.Healthcare'] == False)]

male_average_no_UH = males_no_UH.groupby('Year')['Male Life Expectancy'].mean().reset_index()

print(male_average_no_UH)

plt.plot(male_average_no_UH['Year'], male_average_no_UH['Male Life Expectancy'])
plt.title("Average Male Life Expectancy in Countries without Universal Healthcare (2000-2020)")
plt.xlabel("Year")
plt.ylabel("Average Male Life Expectancy")
plt.show()
```
